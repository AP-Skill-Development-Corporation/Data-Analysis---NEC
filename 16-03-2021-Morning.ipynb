#### Robust Scaler
- If  your dataset contains more no of outliers, then Robust Scaler is best technique for scaling

```python
### Robust Scaler Formula

#     Xi - Q2(X)
#  ---------------
#    Q3(X) - Q1(X)

### Inter Quartile Range (IQR)
#  Q1 --> 25%(Lower Quartile)
#  Q2 --> 50%(Medium Quartile)
#  Q3 --> 75%(Upper Quartile)
```

```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
```

```python
data1.head(20)
```

```python
data1.columns
```

```python
data1.drop(['Unnamed: 0'], axis=1, 
           inplace = True)
```

```python
data1.head()
```

#### Apply Robust Scaler Model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
rs
```

```python
rs2 = rs.fit_transform(data1)
rs2
```

```python
rs2 = pd.DataFrame(rs2, columns = data1.columns)
rs2.head()
```

#### Standard Scaler
- Most commonly used scalling technique. it scales between 0 to mean

```python
### Standard Scaler Formula

#   Xi - mean(X)
# ----------------
#     std(X)
```

```python
# Step1: Import model
# Step2: Initialize model into new object
# Step3: fit the model to dataset
# Step4: Convert array values into dataframe
```

```python
# Import model
from sklearn.preprocessing import StandardScaler
# Initialize model into new object
ss = StandardScaler()
# fit the model to dataset
ss2 = ss.fit_transform(data1)
# Convert array values into dataframe
ss2 = pd.DataFrame(ss2,columns = data1.columns)
```
