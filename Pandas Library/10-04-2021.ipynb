### Different ways to create Series
- List
- Tuples
- Dictionaries
- Numpy array

```python
# Create series using List
s1 = pd.Series([10,11,12,13,14])
s1
```

```python
# create a series using Dictionary
dict3 = {'a':11,'b':'ab','c':12,'d':'de'}
s3 = pd.Series(dict3)
s3
```

```python
import numpy as np

# Create a series using numpy array
s4 = pd.Series(np.array([1,2,3,4,5]))
s4
```

#### Operations between Series (+, -, /, * , ** ) 

```python
s5 = pd.Series([121,1255,899,1000,55])
s6 = pd.Series([234,500,768,1012,2000])
```

```python
s6 = pd.Series([234,500,768,1012,2000], 
               index = ['A','B','C','D','E'])
s6
```

```python
print(s5 + s6)
```

```python
print(s5 ** s6)
```
