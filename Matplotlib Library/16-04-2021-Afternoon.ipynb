### 1. Line plot
- plt.plot()

```python
x_axis = [2,5,8,15]
y_axis = [4,9,16,22]
plt.plot(x_axis, y_axis)
### (2,5), (5,9), ....
```

```python
x_axis = [2,5,8,15]
y_axis = [4,9,16,22]
plt.plot(x_axis, y_axis, label = 'Line Plot')
plt.title("Line Plotting")
plt.xlabel('X_Axis')
plt.ylabel('Y_Axis')
plt.legend()
```

```python
help(plt.legend)
```

```python
x_axis = [2,5,8,15]
y_axis = [4,9,16,22]
plt.plot(x_axis, y_axis, label = 'Line Plot')
plt.legend(loc = 7)
```

```python
import numpy as np
x = np.linspace(0,20,50)
x
```

```python
plt.plot(x, x**2)
plt.plot(x, x**3)
plt.plot(x,x)
```

```python
help(plt.plot)
```

```python
plt.plot(x, x**2, color = 'c', linestyle = '-')
plt.plot(x, x**3, color = 'g', linestyle = '--')
plt.plot(x,x, color = 'm', linestyle = '-.')
```

```python
x2 = np.linspace(10,30,50)
x2
```

```python
plt.plot(x2, np.sin(x2), color = 'green',marker = '+', 
         markersize = 10)
plt.plot(x2, np.cos(x2), color = 'blue', marker = 'd', 
         markersize = 10, markerfacecolor='w')
```
